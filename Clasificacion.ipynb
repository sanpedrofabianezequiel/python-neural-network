```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataCSV = pd.read_csv('data-clasificacion/diario.csv')
```

```python
del dataCSV['number']
```

```python
#Delete colums or fields Emptys
beforeRows =  dataCSV.shape[0]
dataCSV = dataCSV.dropna()
afterRows = dataCSV.shape[0]
print('We lost or deleted rows =',beforeRows - afterRows)
```

```python
#Binarization
cleanData = dataCSV.copy()
#Create a new columns with binarization
cleanData['high_humidity_label'] = (cleanData['relative_humidity_3pm'] > 24.99) * 1
print('Preview cleanData[\'high_humidity_label\']',cleanData['high_humidity_label'])
```

```python
#We filter the copy with the previews column created
target =  cleanData[['high_humidity_label']].copy()
cleanData['relative_humidity_3pm'].head()
```

```python
featuresSelected =  ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        ]
featuresFilteredX = cleanData[featuresSelected].copy()
print('Show preview featuresFilteredX.columns',featuresFilteredX.columns) 
print('Show preview target.columns',target.columns) 
```

```python
#Split data
trainX,testX,trainY,testY = train_test_split(featuresFilteredX,target,test_size=0.33,random_state=324)
#Create clasificator tree
classificationResponse =  DecisionTreeClassifier(max_leaf_nodes=20,random_state=0)
#sent inputs values, (dataX,expectativeY) TRAIN
classificationResponse.fit(trainX,trainY)
#type(classificationResponse)
```

```python
#Test for predicts
predictResponse =  classificationResponse.predict(testX)
predictResponse[:20]
```

```python
#Checked Values manual 
testY['high_humidity_label'][:20]
```

```python
#Calculate the % about the model
accuracy_score(y_true=testY ,y_pred=predictResponse)
```
