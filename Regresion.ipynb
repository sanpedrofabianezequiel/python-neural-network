```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np
plt.style.use("seaborn-dark")
```

```python
x = np.array([0,6.5,13.2,18.1,24.9,28.2,36.1])
z = np.array([22,23,24,21,10,17,8])
y = np.array ([5,10,17,20,25,30,35])
```

```python
#Functions
VARIANT_Y = 5
VARIANT_X = 0.8
def getEstimatedProfit(moneyInvertion):
    return VARIANT_X * moneyInvertion + VARIANT_Y
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x,y,s=60)
plt.ylim(bottom=0)
plt.title('Values')
plt.xlabel('Money to Invertion')
plt.ylabel('Money to expectation')
plt.grid()
plt.show()
```

```python
x_invertion = np.arange(0,40)
y_expectation= 0.8 * x_invertion
```

```python
plt.figure(figsize=(7,7))
plt.plot(x_invertion,y_expectation,color="red",label="Solution")
plt.scatter(x,y,s=60)
plt.ylim(bottom=0)
plt.title('Values')
plt.xlabel('Money to Invertion')
plt.ylabel('Money to expectation')
plt.grid()
plt.show()
```

```python
x_invertion_customer = np.arange(0,40)
y_expectation_customer= getEstimatedProfit(x_invertion_customer)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x_invertion_customer,y_expectation_customer,color="red",label="Solution")
plt.scatter(x,y,s=60)
plt.ylim(bottom=0)
plt.title('Values')
plt.xlabel('Money to Invertion')
plt.ylabel('Money to expectation')
plt.grid()
plt.show()
```
